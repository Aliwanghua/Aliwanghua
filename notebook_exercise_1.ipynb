## Model

```python
import pypsa
pypsa.options.api.new_components_api = True

import pandas as pd

pd.options.plotting.backend = 'plotly'

n = pypsa.Network('exercise_1.xlsx')
```

```python
n.sanitize()
n.optimize(include_objective_constant=True)
```

```python
s = n.statistics

df_curtailment_t = s.curtailment(groupby = False, groupby_time = False, carrier = ['solar','wind']).T
df_curtailment_t.columns = df_curtailment_t.columns.get_level_values("name")

df_curtailment_t.plot()
```

```python
df_curtailment = s.curtailment(groupby=False, carrier=['solar', 'wind']).droplevel("component")
df_curtailment.plot(kind="bar")
```

```python
eb = s.energy_balance(groupby=False).droplevel("component")
eb.plot(kind="barh")
```
